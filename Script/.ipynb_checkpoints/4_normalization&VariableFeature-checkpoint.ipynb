```r
############################### Load libraries  ################################ 
library(Seurat)
library(SeuratDisk)
library(tidyr)
library(stringr)
library(patchwork)
library(ggplot2)


######################### Load scRNA-seq data into R ###########################  


# After generating the count matrices, the next step is data analysis. 
# The R package Seurat is one of the most widely used tools for this purpose. 
# Before creating a Seurat object, several functions are available to read and
# manage different scRNA-seq file formats.




### .mtx file format
obj = ReadMtx(mtx = "/sharedFolder/Data/matrix.mtx.gz",
               cells = "/sharedFolder/Data/barcodes.tsv.gz",
               features = "/sharedFolder/Data/features.tsv.gz")

seurat_obj <- CreateSeuratObject(counts = obj,
                                  project = "myProject",
                                  min.cells = 3,  #filter genes expressed in at least 3 cells
                                  min.features = 200)  # filter cell with less then 200 genes expressed
```

```r

# Show number of cells/genes
ncol(seurat_obj)
nrow(seurat_obj)
```

```r
# After filtering low-quality cells and genes from each sample, the next step 
# is data normalization.
# Sample-specific differences, such as variations in UMI counts that can inflate
# the mean and variance, are properly corrected because normalization is 
# performed independently on each layer.
# Since raw counts are influenced by library size, normalization ensures that 
# expression levels are comparable across cells and variable feature selection 
# is more accurate. 

# show raw counts 
```

```r
# visualize metadata
View(seurat_obj@meta.data)
```

```r
raw_counts <- seurat_obj@assays$RNA
raw_counts=raw_counts@counts
head(raw_counts)
```

```r
# generate histogram for raw counts 
raw_hist <- hist(as.vector(raw_counts), breaks = 50, plot = FALSE)
# log-transform the frequency (y-axis)
raw_hist$counts <- log1p(raw_hist$counts)  
# plot raw counts histogram with log-transformed frequency
png("/sharedFolder/raw_counts_hist.png",
       width = 1500, height = 1500,
       res = 300, units = "px")
plot(raw_hist, col = "#bc91a5", 
     main = NULL,
     xlab = "Raw counts", 
     ylab = "Log-transformed frequency")
dev.off()
```

```r

# Seurat normalizes gene expression by dividing each gene count by the total 
# UMI counts per cell, multiplying by a scale factor (1e4 by default), and 
# log-transforming the result (global-scaling):
# - NOTE: log1p computes ln(1 + x)
# - NOTE: Raw counts are stored in the layer named 'counts' and normalized
# counts are stored in the layer named 'data'.  
```

```r
seurat_obj <- NormalizeData(seurat_obj, 
                                   normalization.method = "LogNormalize", 
                                   scale.factor = 10000)
```

```r
normalized_counts <- seurat_obj@assays$RNA
normalized_counts=normalized_counts@data
```

```r

# generate histogram for log-normalized counts
normalized_hist <- hist(as.vector(normalized_counts), breaks = 50, plot = FALSE)
# log-transform the frequency (y-axis)
normalized_hist$counts <- log1p(normalized_hist$counts)  
# plot log-normalized counts histogram with log-transformed frequency
png("/sharedFolder/normalized_counts_hist.png",
    width = 1500, height = 1500,
    res = 300, units = "px")
plot(normalized_hist, col = "#bc91a5", 
     main = NULL,
     xlab = "Log-normalized counts", 
     ylab = "Log-transformed frequency")
dev.off()
```

```r
#  Alternatively, we can compute relative counts using a scaling factor of 1e6
# (counts per million or CPM):
cpm.seurat.obj <- NormalizeData(seurat_obj, 
                                normalization.method = "RC", 
                                scale.factor = 1e6)

# Same as:
# total_counts <- colSums(merged.seurat.obj@assays$RNA$counts.1)
# head(t(t(merged.seurat.obj@assays$RNA$counts.1) / total_counts * 1e6))

# show CPM
cpm_counts <- cpm.seurat.obj@assays$RNA
cpm_counts=cpm_counts@data
head(cpm_counts)

# generate histogram for CPM 
cpm_hist <- hist(as.vector(cpm_counts), breaks = 50, plot = FALSE)
# log-transform the frequency (y-axis)
cpm_hist$counts <- log1p(cpm_hist$counts)  
# plot CPM histogram with log-transformed frequency
png("/sharedFolder/cpm_counts_hist.png",
    width = 1500, height = 1500,
    res = 300, units = "px")
plot(cpm_hist, col = "#bc91a5", 
     main = NULL,
     xlab = "Counts per million (CPM)", 
     ylab = "Log-transformed frequency")
dev.off()
```

```r
########################## Identify variable genes #############################

# We calculate the standardized variance of gene expression across cells for 
# each gene. This is essential for identifying genes that vary the most across 
# cells (i.e., genes with high expression in some cells and low in others),
# which are often the most biologically relevant for downstream analyses 
# like PCA or clustering.

# - NOTE: When identifying variable features, Seurat first calculates them for 
# each individual layer/sample, then identifies the common ones across layers 
# (consistently variable across all samples), and finally, it extends the list
# of variable features until the desired number is reached (2000 by default).
seurat_obj <- FindVariableFeatures(seurat_obj, 
                                          selection.method = "vst", 
                                          nfeatures = 2000)

# visualize metadata associated with genes
metadata=seurat_obj@assays$RNA
metadata=metadata@meta.features
View(metadata)

# plot the top 10 variable genes 
top_10_genes <- head(VariableFeatures(seurat_obj), 10)
top_10_genes_plot <- VariableFeaturePlot(seurat_obj)
LabelPoints(plot = top_10_genes_plot, points = top_10_genes, 
            repel = TRUE, xnudge = 0, ynudge = 0)
ggsave("/sharedFolder/top_10_genes.png",
       width = 2000, height = 2000,
       dpi = 300, units = "px")

# We can observe that most of the highly variables genes (IGHG1, IGHA1, IGJ, IGHA2)
# are immunoglobulin genes, which are involved in the immune response.
# We can investigate later in which cells they are expressed (e.g., B cells). 


# save result 
saveRDS(seurat_obj, "/sharedFolder/example.rds")

```
